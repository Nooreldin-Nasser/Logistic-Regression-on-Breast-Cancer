## Normalize the Data

```python
x_normalized = (x_data - x_data.min()) / (x_data.max() - x_data.min())
x_normalized
```

## Split the Data

```python
train_ratio = 0.7
train_size = int(len(x_normalized)*train_ratio)

x_train = np.array(x_normalized[:train_size])
x_test = np.array(x_normalized[train_size:])

y_train = np.array(y_data[:train_size])
y_test = np.array(y_data[train_size:])

```

## Build Our Model

```python
def sigmoid(z):
    g = 1 / (1+np.exp(-z))
    return g
```

```python
def compute_cost_logistic(x,y,w,b):

    m = x.shape[0]
    cost = 0.0
    for i in range (m):
        f_wb = sigmoid(np.dot(x[i],w) + b)
        cost += - y[i]*(np.log(f_wb)) - (1 - y[i]) * (np.log(1-f_wb))
    cost = cost / m
    return cost
```

```python
def compute_gradient_logistic(x,y,w,b):

    m,n =x.shape

    dj_dw = np.zeros(n)
    dj_db = 0.0

    for i in range (m):
        f_wb = sigmoid(np.dot(x[i],w) + b)
        err = f_wb - y[i]

        for j in range(n):
            dj_dw += err * x[i,j]
        dj_db += err

    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_dw , dj_db 
```

```python
X_tmp = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_tmp = np.array([0, 0, 0, 1, 1, 1])
w_tmp = np.array([7.,8.])
b_tmp = 1.
dj_dw_tmp, dj_db_tmp = compute_gradient_logistic(X_tmp, y_tmp, w_tmp, b_tmp)
print(f"dj_db: {dj_db_tmp}" )
print(f"dj_dw: {dj_dw_tmp.tolist()}" )
```

```python
def gradient_logistic(x,y,w_in,b_in,cost_func,grad_func,alpha,num_iters):
    J_history = []
    w = w_in  #avoid modifying global w within function
    b = b_in
    for i in range(num_iters):
        dj_dw , dj_db = grad_func(x,y,w,b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        J_history.append(cost_func(x,y,w,b))
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]}   ") 
    
    return w , b ,J_history
```

```python
w_tmp  = np.zeros_like(x_train[0])
b_tmp  = 0.
alph = 0.5
iters = 10000

w_out, b_out, _ = gradient_logistic(x_train, y_train, w_tmp, b_tmp,compute_cost_logistic,compute_gradient_logistic, alph, iters) 
print(f"\nupdated parameters: w:{w_out}, b:{b_out}")
```

```python
def prediction(x,y,w,b):
    m,_ = x.shape
    p = []
    for i in range(m):
        p.append((sigmoid(np.dot(x[i], w) + b)) >= 0.6)

        print(f"prediction: {p[-1]:0.2f}, target value: {y[i]}")
    return p
```

```python
p = np.array(prediction(x_test,y_test,w_out,b_out))
p.shape
```

```python
print('Train Accuracy: %f'%(np.mean(p == y_test) * 100))
```
